# 擴展縮放

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('Chimei_Museum.jpg')
img = cv2.resize(img,dsize = None,fx=0.8, fy=0.9, interpolation=cv2.INTER_CUBIC)

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
img.shape
```

# 平移

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
img = cv2.imread('Chimei_Museum.jpg' ,0)
rows, cols = img.shape
rows, cols
M = np.float32([[1, 0, 100], [0, 1, 100]])
dst = cv2.warpAffine(img, M, (cols, rows),borderValue=(255,255,255))

dst = cv2.cvtColor(dst, cv2.COLOR_BGR2RGB)
plt.imshow(dst)
```


# 旋轉

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
img = cv2.imread('Chimei_Museum.jpg')
rows, cols, ch = img.shape

M = cv2.getRotationMatrix2D((0 , 0), -10, 1)

dst = cv2.warpAffine(img, M, (2 * cols, 2 * rows),borderValue=(255,255,255))

dst = cv2.cvtColor(dst, cv2.COLOR_BGR2RGB)
plt.imshow(dst)
```

# 仿射變換

```python
import cv2 
import numpy as np 
from matplotlib import pyplot as plt

img=cv2.imread('drawing.png') 
rows,cols,ch=img.shape
pts1=np.float32([[50,50],[200,50],[50,200]]) 
pts2=np.float32([[10,100],[200,50],[100,250]])
M=cv2.getAffineTransform(pts1,pts2)
dst=cv2.warpAffine(img,M,(cols,rows))

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.subplot(121), plt.imshow(img), plt.title('Input')
dst = cv2.cvtColor(dst, cv2.COLOR_BGR2RGB)
plt.subplot(122), plt.imshow(dst), plt.title('Output')
plt.show()
```

# 透視變換

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('sudoku.JPG')
rows, cols, ch = img.shape

pts1 = np.float32([[56,65],[368,52],[28,387],[389,390]]) 
pts2 = np.float32([[0,0],[300,0],[0,300],[300,300]])
M=cv2.getPerspectiveTransform(pts1,pts2)
dst=cv2.warpPerspective(img,M,(300,300))


img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.subplot(121), plt.imshow(img), plt.title('Input')
dst = cv2.cvtColor(dst, cv2.COLOR_BGR2RGB)
plt.subplot(122), plt.imshow(dst), plt.title('Output')
plt.show()
```
