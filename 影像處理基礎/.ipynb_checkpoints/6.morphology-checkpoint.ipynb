# 腐蝕 (erode)

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('j.png', 0)
kernel = np.ones((5,5), np.uint8)
# print(kernel)

erorsion_img = cv2.erode(img, kernel, iterations=1)

# img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
# plt.imshow(img)
# erorsion_img = cv2.cvtColor(erorsion_img, cv2.COLOR_BGR2RGB)
# plt.imshow(erorsion_img)

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.subplot(121), plt.imshow(img), plt.title('img')
plt.xticks([]), plt.yticks([])
erorsion_img = cv2.cvtColor(erorsion_img, cv2.COLOR_BGR2RGB)
plt.subplot(122), plt.imshow(erorsion_img), plt.title('erorsion_img')
plt.xticks([]), plt.yticks([])
plt.show()
```

# 膨脹 (dilate)

```python
import cv2
import numpy as np
img = cv2.imread('j.png', 0)
kernel = np.ones((5,5), np.uint8) 
dilation_img = cv2.dilate(img, kernel, iterations=1)

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.subplot(121), plt.imshow(img), plt.title('img')
plt.xticks([]), plt.yticks([])
dilation_img = cv2.cvtColor(dilation_img, cv2.COLOR_BGR2RGB)
plt.subplot(122), plt.imshow(dilation_img), plt.title('dilation_img')
plt.xticks([]), plt.yticks([])
plt.show()
```

# Opening：先腐蝕再膨脹

```python
import cv2
import numpy as np
img = cv2.imread("j_o.jpg")

kernel = np.ones((5,5), np.uint8)
opening_img = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel,iterations=1)

# opening_img = cv2.cvtColor(opening_img, cv2.COLOR_BGR2RGB)
# plt.imshow(opening_img)
# cv2.imwrite('j_opening.jpg', opening_img)

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.subplot(121), plt.imshow(img), plt.title('img')
plt.xticks([]), plt.yticks([])
opening_img = cv2.cvtColor(opening_img, cv2.COLOR_BGR2RGB)
plt.subplot(122), plt.imshow(opening_img), plt.title('opening_img')
plt.xticks([]), plt.yticks([])
plt.show()
```

# Closing：先膨脹再腐蝕

# gradient  = 圖像膨脹 - 圖像腐蝕
